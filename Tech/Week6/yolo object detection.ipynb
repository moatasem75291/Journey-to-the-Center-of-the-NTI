```python
from ultralytics import YOLO

model = YOLO('yolov10n.pt')

training_params = {
    'data': '/kaggle/input/pinotnoirgrapes/data.yaml',  # Path to your data.yaml file
    'epochs': 50,            # Increase number of epochs for more training
    'imgsz': 1280,            # Image size; you may increase if your hardware allows
    'batch': 8,               # Adjust batch size depending on your GPU memory
    'optimizer': 'Adam',      # Try Adam optimizer
    'lr0': 0.001,             # Initial learning rate, you may adjust this
    'lrf': 0.1,               # Final learning rate
    'momentum': 0.937,        # Momentum for SGD
    'weight_decay': 0.0005,   # Regularization
    'hsv_h': 0.015,           # Hue augmentation
    'hsv_s': 0.7,             # Saturation augmentation
    'hsv_v': 0.4,             # Value augmentation
    'flipud': 0.5,            # Vertical flip
    'fliplr': 0.5,            # Horizontal flip
    'scale': 0.5,             # Scale augmentation
    'translate': 0.1,         # Translate augmentation
    'mixup': 0.2              # Mixup augmentation
}

history = model.train(**training_params)
history = model.train(data='/kaggle/input/pinotnoirgrapes/data.yaml', epochs=50, imgsz=1280)
```

```python
results = model.predict(source='/kaggle/input/pinotnoirgrapes/Grapes/Grapes/PinotNoir/test/images', save=True)
```

```python
metrics = model.val()

# Print the main metrics
print(f"mAP@0.5: {metrics.box.map50:.3f}")
print(f"mAP@0.5:0.95: {metrics.box.map:.3f}")
print(f"Mean Precision: {metrics.box.mp:.3f}")
print(f"Mean Recall: {metrics.box.mr:.3f}")
print(f"F1-score (calculated as harmonic mean of Precision and Recall): {2 * metrics.box.mp * metrics.box.mr / (metrics.box.mp + metrics.box.mr):.3f}")
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import cv2

img_path = '/kaggle/input/pinotnoirgrapes/Grapes/Grapes/PinotNoir/test/images/IMG_0241_0.jpg'
results = model(img_path)

boxes = results[0].boxes
print(f"Detected {len(boxes)} grape clusters.")

image = cv2.imread(img_path)

for box in boxes:
    x1, y1, x2, y2 = map(int, box.xyxy[0])
    cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 4)

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
import cv2
from ultralytics import YOLO
import time


video_path = '/kaggle/input/pinotnoirgrapes/Grapes/Grapes/videos/grape_track_count.avi'
output_video_path = 'grape_detection_output.avi'  # Path for the output video

def process_video(video_path, output_video_path):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"Error opening video file: {video_path}")
        return

    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    
    fourcc = cv2.VideoWriter_fourcc(*'XVID')  # Codec for saving the video
    out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

    frame_count = 0
    start_time = time.time()

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break  # Exit if no more frames

        results = model(frame)

        boxes = results[0].boxes

        for box in boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # x1, y1, x2, y2 are the corners of the box
            confidence = box.conf[0]  # Confidence score for the detection
            # Draw the box with confidence score
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 5)
            cv2.putText(frame, f"Grape {confidence:.2f}", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

        out.write(frame)

        frame_count += 1
        elapsed_time = time.time() - start_time
        fps_calculated = frame_count / elapsed_time
        print(f"Processed frame {frame_count} - FPS: {fps_calculated:.2f}")

    cap.release()
    out.release()
    print(f"Video saved as {output_video_path}")

process_video(video_path, output_video_path)
```
