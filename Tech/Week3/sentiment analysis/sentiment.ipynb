```python
def mapping_binary_columns(df, column_names):
    for column_name in column_names:
        if column_name == 'gender':
            df[column_name] = df[column_name].map({'M': 1, 'F': 0})
        else:
            df[column_name] = df[column_name].map({'Y': 1, 'N': 0})
    return df
columns_to_map = ["gender", "has_car", "has_property"]
train = mapping_binary_columns(train, columns_to_map)
test = mapping_binary_columns(test, columns_to_map)
```

```python
def transform_age(df):
    df['age'] = (np.abs(df['age']) / 365).astype(int)
    df['account_age'] = (np.abs(df['age'])).astype(int)
    return df
train = transform_age(train)
test = transform_age(test)
```
